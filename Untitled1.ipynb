```python
import pandas as pd
import torch
from transformers import (
    MarianMTModel, MarianTokenizer,
    T5ForConditionalGeneration, T5Tokenizer,
    MBartForConditionalGeneration, MBart50TokenizerFast,
    BertTokenizer, BertModel
)
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
```

```python
df = pd.read_csv("/content/parallel_sentences_ophtalmology_test.csv")
sample_texts = df['en'].head(50).tolist()
actual_translations = df['ru'].head(50).tolist()
```

```python
# Evaluation Function
def word_level_accuracy(predictions, references):
    total_words, correct_words = 0, 0
    for pred, ref in zip(predictions, references):
        pred_words = set(pred.strip().split())
        ref_words = set(ref.strip().split())
        correct_words += len(pred_words & ref_words)
        total_words += len(ref_words)
    return (correct_words / total_words) * 100 if total_words > 0 else 0
```

```python
# 2. MarianMT Model (EN -> RU)
marian_model_name = "Helsinki-NLP/opus-mt-en-ru"
marian_tokenizer = MarianTokenizer.from_pretrained(marian_model_name)
marian_model = MarianMTModel.from_pretrained(marian_model_name)

def translate_marian(texts):
    inputs = marian_tokenizer(texts, return_tensors="pt", padding=True, truncation=True)
    outputs = marian_model.generate(**inputs)
    return [marian_tokenizer.decode(t, skip_special_tokens=True) for t in outputs]
```

```python
# 3. T5 for EN -> RU
t5_model_name = "t5-small"
t5_tokenizer = T5Tokenizer.from_pretrained(t5_model_name)
t5_model = T5ForConditionalGeneration.from_pretrained(t5_model_name)

def translate_t5(texts):
    inputs = t5_tokenizer(["translate English to Russian: " + text for text in texts],
                          return_tensors="pt", padding=True, truncation=True)
    outputs = t5_model.generate(**inputs)
    return [t5_tokenizer.decode(t, skip_special_tokens=True) for t in outputs]
```

```python
# 4. mBART for EN -> RU
mbart_model_name = "facebook/mbart-large-50-many-to-many-mmt"
mbart_tokenizer = MBart50TokenizerFast.from_pretrained(mbart_model_name)
mbart_model = MBartForConditionalGeneration.from_pretrained(mbart_model_name)

def translate_mbart(texts):
    mbart_tokenizer.src_lang = "en_XX"
    inputs = mbart_tokenizer(texts, return_tensors="pt", padding=True, truncation=True)
    outputs = mbart_model.generate(**inputs, forced_bos_token_id=mbart_tokenizer.lang_code_to_id["ru_RU"])
    return [mbart_tokenizer.decode(t, skip_special_tokens=True) for t in outputs]
```

```python
# 5. BERT Embeddings (Demo for contextual understanding)
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
bert_model = BertModel.from_pretrained("bert-base-uncased")
with torch.no_grad():
    inputs = bert_tokenizer(sample_texts, return_tensors="pt", padding=True, truncation=True)
    outputs = bert_model(**inputs)
    print("\n[BERT] Embedding shape:", outputs.last_hidden_state.shape)  # (batch_size, seq_len, hidden_dim)
```

```python
# 6. LSTM Demo (Just structure and shape output)
class SimpleLSTM(nn.Module):
    def __init__(self, input_dim=768, hidden_dim=256, num_layers=1):
        super(SimpleLSTM, self).__init__()
        self.lstm = nn.LSTM(input_dim, hidden_dim, num_layers, batch_first=True)

    def forward(self, x):
        out, _ = self.lstm(x)  # No error now
        return out

lstm_model = SimpleLSTM()
dummy_input = torch.randn(10, 20, 768)  # batch_size=10, seq_len=20
lstm_output = lstm_model(dummy_input)
print("[LSTM] Output Shape:", lstm_output.shape)
```

```python
t5_out = translate_t5(sample_texts)
t5_accuracy = word_level_accuracy(t5_out, actual_translations)
print(f"[Model] T5 Accuracy: {t5_accuracy * 100:.2f}%")
```
